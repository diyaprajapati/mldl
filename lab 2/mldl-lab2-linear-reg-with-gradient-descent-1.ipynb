```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/age-data/age data.csv")
```

```python
X= data["Age"]
y= data["Weight"]
X_n = np.array(X)
X_r = X_n.reshape(1,-1)
y_n = np.array(y)
y_r = y_n.reshape(1,-1)
X_r
```

```python
plt.scatter(X_r,y_r, c= 'r')
```

```python
m = 0
c= 0
L = 0.001
epochs=1000
n= float(len(X))
#y.shape
```

```python
for i in range(10): 
    y_pred = m*X_r + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X_r * (y_r - y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(y_r - y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    
    c = c - L * D_c  # Update c
    plt.scatter(X_r,y_r, c= 'r')
    
#print (m, c)
plt.plot(X_r,y_pred)
```
